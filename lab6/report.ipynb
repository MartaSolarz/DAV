### Task 1) Analyze the dataset: [ans.csv](ans.csv)

Calculate the missing values (fill the tables).

| Property | Value | Accuracy |
| --- | --- | --- |
| Mean of x | ? | ? |
| Sample variance of x | ? | ? |
| Mean of y | ? | ? |
| Sample variance of y | ? | ? |
| Correlation between x and y | ? | ? |
| Linear regression line (y = a+bx)| ? | ? |
| Coefficient of determination of the linear regression| ? | ? |


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv("data/ans.csv", index_col=0)

# Calculate the mean of x and y
mean_x = df.iloc[:, :4].mean().mean()
mean_y = df.iloc[:, 4:].mean().mean()

# Calculate the sample variance of x and y
var_x = df.iloc[:, :4].values.flatten().var(ddof=1)
var_y = df.iloc[:, 4:].values.flatten().var(ddof=1)
 
# Calculate the correlation between x and y
correlation = np.corrcoef(df.iloc[:, :4].values.flatten(), df.iloc[:, 4:].values.flatten())[0, 1]
 
# Calculate the linear regression line (y = a+bx) and Step 7: Calculate the coefficient of determination
x = df.iloc[:, :4].values.flatten()
y = df.iloc[:, 4:].values.flatten()

model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# Coefficients of the linear regression line
a = model.intercept_
b = model.coef_[0]
 
# Coefficient of determination (R^2)
r_squared = model.score(x.reshape(-1, 1), y)
```

```python
# Results
{
    "mean_x": mean_x,
    "var_x": var_x,
    "mean_y": mean_y,
    "var_y": var_y,
    "correlation_xy": correlation,
    "linear_regression_line": f"y = {a} + {b}x",
    "coefficient_of_determination": r_squared
}
```

| Property | Value            | Accuracy                          |
| --- |------------------|-----------------------------------|
| Mean of x | 9.00             | Exact                             |
| Sample variance of x | 10.23            | Approximate (up to 2 places)      |
| Mean of y | 7.50             | Approximate (up to 2 places)      |
| Sample variance of y | 3.84             | Approximate  (up to 2 places)     |
| Correlation between x and y | 0.816            | Approximate (up to 3 places)      |
| Linear regression line (y = a+bx) | y = 3.00 + 0.50x | Approximate    (up to 2 places)   |
| Coefficient of determination of the linear regression | 0.67             | Approximate      (up to 2 places) |


Scatter plot for Task 1

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0,0].scatter(df.x1, df.y1, label="Dataset 1", color="red", alpha=0.5)
ax[0,0].plot(x, model.predict(x.reshape(-1, 1)), color="blue")
ax[0,0].set_xlabel("x1")
ax[0,0].set_ylabel("y1")
ax[0,0].title.set_text("Dataset 1")

ax[0,1].scatter(df.x2, df.y2, label="Dataset 2", color="red", alpha=0.5)
ax[0,1].plot(x, model.predict(x.reshape(-1, 1)), color="blue")
ax[0,1].set_xlabel("x2")
ax[0,1].set_ylabel("y2")
ax[0,1].title.set_text("Dataset 2")

ax[1,0].scatter(df.x3, df.y3, label="Dataset 3", color="red", alpha=0.5)
ax[1,0].plot(x, model.predict(x.reshape(-1, 1)), color="blue")
ax[1,0].set_xlabel("x3")
ax[1,0].set_ylabel("y3")
ax[1,0].title.set_text("Dataset 3")

ax[1,1].scatter(df.x4, df.y4, label="Dataset 4", color="red", alpha=0.5)
ax[1,1].plot(x, model.predict(x.reshape(-1, 1)), color="blue")
ax[1,1].set_xlabel("x4")
ax[1,1].set_ylabel("y4")
ax[1,1].title.set_text("Dataset 4")

plt.tight_layout()
plt.show()
```
